```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
  
# Creating an instance
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

from selenium.webdriver.common.by import By
driver.get("https://wrlc-gm.primo.exlibrisgroup.com/discovery/search?query=creator,exact,Linton%20Wells,AND&tab=Everything&search_scope=MyInst_and_CI&sortby=rank&vid=01WRLC_GML:01WRLC_GML&lang=en&mode=advanced&offset=0")
```

```python
src = driver.page_source
  
   # Now using beautiful soup
soup = BeautifulSoup(src, 'lxml')
results = soup.find("div", {"id": "searchResultsContainer"})
print(results)
```

```python
for link in results.find_all("h3", {"class": "item-title"}):
    data = link.find('a').get('href')
    name = link.get_text()
    print(name)
    print(data.split(" ")[0])
        
```

```python
temp = soup.find("div", {"class": "search-toolbar-title layout-row"})
count = temp.find("span", {"class": "results-count"})
print(count.get_text().split('  ')[0])
```
